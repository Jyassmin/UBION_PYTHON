#### 2. 파생변수 만들기 SQL & Pandas

- SQL

```python
sql = """
    select *
    from emp
"""
cursor.execute(sql)
result = cursor.fetchall()
print(pd.DataFrame(result))
```

- Pandas

```python
# 파생변수 만들기 SQl vs Pandas
sql = """
    select *
    from emp
"""
cursor.execute(sql)
result = cursor.fetchall()
df = pd.DataFrame(result)
```

```python
##apply함수를 이용하여 - lambda 함수 사용
df2 = df.copy()
df2["job_code"] = df2["JOB"].apply(lambda x : "1" if (x == "SALESMAN") else "2" if (x == "MANAGER") else "3")
df2[["EMPNO", "ENAME", "JOB", "job_code"]]
```

```python
#일반함수 사용
def change(x):
    if x == "SALESMAN":
        return "1"
    elif x == "MANAGER":
        return "2"
    else:
        return "3"

df3 = df.copy()
df3["job_code"] = df3["JOB"].apply(change)
df3[["EMPNO", "ENAME", "JOB", "job_code"]]
```

###  3. 값 변경 + 조회

sql로

```python
# replace사용! job컬럼에서 "SALESMANS"을 "SALES"로 바꾼다!
sql = """
    select JOB, replace(JOB, "SALESMAN", "SALES") as JOB_
    from emp
"""
cursor.execute(sql)
result = cursor.fetchall()
pd.DataFrame(result)
```

pandas로

```python
# pandas에서 위의 sql 출력과 같이
# replace사용! job컬럼에서 "SALESMAN"을 "SALES"로 바꾼다!
sql = """
    select *
    from emp
"""
cursor.execute(sql)
result = cursor.fetchall()
df = pd.DataFrame(result)
df
```

```python
### 방법1: 일반함수

def change_sales(x):
    if x == "SALESMAN":
        return "SALES"
    else:
        return x

df4 = df.copy()
df4["JOB_"] = df4["JOB"].apply(change_sales)
df4[["JOB", "JOB_"]]
```

```python
### 방법2: replace

df4 = df.copy()
df4["JOB_"] = df4["JOB"].replace("SALESMAN", "SALES")
df4[["JOB", "JOB_"]]
```

## 5. Q
    - sql에서 if문을 사용하여 새로운 파생변수 생성
    - SAL이 2000이상이면 High 미만이면 Low
    - 컬럼의 이름은 H/L

```python
#sql
sql = """
    select SAL, if(SAL >= 2000, "High", "Low") as `H/L`
    from emp
"""
db.executeall(sql)
```

```python
#pandas
def change(x):
    if x >= 2000:
        return "High"
    else:
        return "Low"

df6 = df.copy()
df6["H/L"] = df6["SAL"].apply(change)
df6[["SAL", "H/L"]]
```
