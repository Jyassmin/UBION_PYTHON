## subplot

```python
x = [1,2,3,4]
y = [1,3,4,9]

plt.plot(x,y, "--r")    # "--r"은 점선
plt.plot(x, [4,2,6,4], "oy")    # "oy"는 노랑점
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Class & Count")  #제목
plt.legend(["A","B"], loc=)    #legend는 범례추가. loc는 위치지정
plt.show
```

```python
# subplot(a,b,c) -> a행, b열로 공간을 나눈 뒤 c번째 공간을 쓰자.
plt.subplot(1,2,1)
plt.plot([1,2,3,4])

plt.subplot(1,2,2)
plt.plot([2,4,6,8])
plt.show()
```

```python
# Q. 임의의 그래프 4개를 2x2 subplot출력
plt.subplot(2,2,1)
plt.plot([1,1,2,2])
plt.legend(["A"])   #범례

plt.subplot(2,2,2)
plt.plot([1,1,2,2])

plt.subplot(2,2,3)
plt.plot([1,1,2,2])

plt.subplot(2,2,4)
# plt.plot([1,1,2,2])
plt.bar([1,2,3,4],[2,2,1,1])   #bar차트 추가
plt.show()
```

```python
x = [1,2,3,4]
y= [80,40,70,40]

plt.bar(x,y)
plt.show
```

```python
plt.bar(x,y, color=["r","y","b"], width=0.3)    #색상은 돌아가면서 쓰임
plt.show()
```

```python
plt.bar(x,y, color=["r","y","b"], width=0.3)    #색상은 돌아가면서 쓰임
plt.xticks(x, ["maths", "seience", "history", "english"])   #xticks는 대체해서 보여주는 index역할
plt.yticks(y, ["maths", "seience", "history", "english"])
plt.show()
```

> ## scatter, 산점도

```python
import numpy as np
```

```python
#산점도
x = np.random.rand(50)  #랜덤하게 값 가져옴
y = np.random.rand(50)

plt.scatter(x,y)
plt.show()
```

- 산점도 그래프 옵션 추가(영역확장, 색상변경, 투명도 조절

```python
color = np.random.rand(50)
area = (20*np.random.rand(50))**2   #점끼리 크기차이를 벌려주기 위해 상수를 곱하고 제곱해줌

plt.scatter(x,y, s=area, c=color, alpha=1)  #alpha는 투명도를 의미
plt.show()
```
